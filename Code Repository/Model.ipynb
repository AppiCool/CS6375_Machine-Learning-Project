# Data Transformation

```python
dummy = df.dropna()
dummy
```

##### Date and time column transformation


```python
dates = pd.to_datetime(dummy.Date_of_Journey)
dates
```

```python
#dummy['start_year']=pd.DatetimeIndex(dates).year
dummy['dep_month']=pd.DatetimeIndex(dates).month
dummy['dep_day']=pd.DatetimeIndex(dates).day
dummy['Date_of_Journey']=dates
dummy
```

```python
pd.DatetimeIndex(pd.to_datetime(dummy['Arrival_Time'])).hour
```

```python
dummy['Dep_hour']=pd.DatetimeIndex(pd.to_datetime(dummy['Dep_Time'])).hour
dummy['Dep_minutes']=pd.DatetimeIndex(pd.to_datetime(dummy['Dep_Time'])).minute
dummy['Arr_hour']=pd.DatetimeIndex(pd.to_datetime(dummy['Arrival_Time'])).hour
dummy['Arr_minutes']=pd.DatetimeIndex(pd.to_datetime(dummy['Arrival_Time'])).minute
dummy.head()
```

```python
dummy.columns
```

##### Converting duration to minutes

```python
def process_duration(test):
    ans = 0
    for i in test.split(' '):
        if 'h' in i:
            ans+=int(i.split('h')[0]) * 60
        elif 'm' in i:
            ans+=int(i.split('m')[0]) * 1
    return ans

process_duration('2h 30m')
```

```python
dummy['Duration']=dummy['Duration'].apply(process_duration)
dummy
```

##### Dropping unecessary columns

```python
dummy = dummy.drop(columns=['Date_of_Journey','Route','Dep_Time','Arrival_Time','Additional_Info'])
dummy
```

##### Label Encoding

```python
encoders = {}
for i in ['Airline','Source','Destination']:
    le = LabelEncoder()
    dummy[i]=le.fit_transform(dummy[i])
    encoders[i]=le
dummy.head()
```

```python
stops = {'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4}
def encode_stops(arr):
    return stops[arr]
dummy['Total_Stops'] = dummy['Total_Stops'].apply(encode_stops)
```

```python
dummy.head()
```

```python
dummy.to_csv('refined.csv')
```

# Data split

```python
x = dummy.drop(columns='Price')
y = dummy['Price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(
                                        x,
                                        y,
                                        random_state=1234,test_size = 0.20,
                                        shuffle=True
                                        )
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Model building

```python
print('--XGBoost')
xgb = XGBRegressor(random_state=1234) 
xgb.fit(x_train,y_train)
print('Train Score:',xgb.score(x_train,y_train))
print('Test Score:',xgb.score(x_test,y_test))
print()
```

```python
xgb.get_params()
```

```python
print('--XGBoost RF')
xgbrf = XGBRFRegressor(random_state=1234) 
xgbrf.fit(x_train,y_train)
print('Train Score:',xgbrf.score(x_train,y_train))
print('Test Score:',xgbrf.score(x_test,y_test))
print()
```

```python
xgbrf.get_params()
```

```python


from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# assume x_train, y_train, x_test, y_test are your data
xgbrf = XGBRFRegressor(random_state=1234)


# predict the probabilities for the test set
probs = xgbrf.predict(x_test)
print(probs)

# calculate the false positive rate and true positive rate for different threshold values
fpr, tpr, thresholds = roc_curve(y_test, probs)

# calculate the area under the curve
roc_auc = auc(fpr, tpr)

# plot the ROC curve
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')  # diagonal line for random guessing
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
    
   
```
